```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the dataset
df = pd.read_csv(r"C:\Users\eissa\dd\Full_clustered_Data.csv", encoding="ascii")
```

```python
# Calculate the standard deviation for the specified columns grouped by the 'Target' column
std_dev = df.groupby('Target')[['hip_knee_angle_r', 'right_shoulder_elbow_angle']].std()

std_dev
```

```python
# Plotting
std_dev.plot(kind='bar', figsize=(10, 6))
plt.title('Standard Deviation of Angles by Target Cluster')
plt.ylabel('Standard Deviation')
plt.xlabel('Target Cluster')
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Separate the dataframe by cluster
df_0 = df[df['Target'] == 0]
df_1 = df[df['Target'] == 1]
df_2 = df[df['Target'] == 2]

# Plot for each cluster
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

clusters = [df_0, df_1, df_2]
for i, cluster_df in enumerate(clusters):
    axs[i].plot(cluster_df.index, cluster_df['right_shoulder_elbow_angle'])
    axs[i].set_title('Cluster ' + str(i))
    axs[i].set_xlabel('Index')
    axs[i].set_ylabel('Right Shoulder Elbow Angle')
    axs[i].set_ylim([0, 180])

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Full_clustered_Data.csv", encoding="ascii")

# Define a range for slicing the data to visualize a portion of it
start_index = 0
end_index = 2000

# Separate and slice the dataframe by cluster
sliced_df_0 = df[df['Target'] == 0].iloc[start_index:end_index]
sliced_df_1 = df[df['Target'] == 1].iloc[start_index:end_index]
sliced_df_2 = df[df['Target'] == 2].iloc[start_index:end_index]

# Plot for each sliced cluster
fig, axs = plt.subplots(3, 1, figsize=(10, 15))

clusters = [sliced_df_0, sliced_df_1, sliced_df_2]
for i, cluster_df in enumerate(clusters):
    axs[i].plot(cluster_df.index, cluster_df['right_shoulder_elbow_angle'])
    axs[i].set_title('Cluster ' + str(i))
    axs[i].set_xlabel('Frame(sec)')
    axs[i].set_ylabel('Right Shoulder Elbow Angle')
    axs[i].set_ylim([0, 180])

plt.tight_layout()
plt.show()
```
